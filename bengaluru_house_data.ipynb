```python
def find_average(value):
    conversion_factors = {
        'Acres': 4046.86,
        'Cents': 40.4686,
        'Guntha': 101.171,
        'Grounds': 203.0,
        'Sq. Meter': 1.0,
        'Sq. Yards': 0.836127,
        'Sq. Feet': 0.092903
    }
    
   
    if " - " in value:
        parts = value.split(" - ")
        return (float(parts[0].strip()) + float(parts[1].strip())) / 2
    
    
    for unit, factor in conversion_factors.items():
        if unit in value:
            numeric_part = float(value.replace(unit, '').strip())
            return numeric_part * factor
    
    
    if value.replace('.', '', 1).isdigit():
        return float(value)
    
    raise ValueError(f"Unexpected format: {value}")


df1['total_sqft_standard'] = df1.total_sqft.apply(find_average)


 
```
